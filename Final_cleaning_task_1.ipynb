```python
import pandas as pd
import numpy as np
```

```python
final=pd.read_csv('https://raw.githubusercontent.com/biukaC/DA-Internship-ProgUA/refs/heads/main/copy.csv')
cleaned=final.copy()
cleaned.head()
```

```python
print(cleaned.shape)
```

```python
cleaned.columns
```

```python
cleaned['Lead sale ₴'].unique()
```

```python
cleaned['Lead sale ₴'] = cleaned['Lead sale ₴'].astype('float64')
```

```python
cleaned['phone'].unique()
```

```python
cleaned['phone'] = (cleaned['phone']
               .fillna('')
               .astype(str)
               .str.replace(r'(?!^\+)[^\d]', '', regex=True)
               .str.strip())

print(cleaned['phone'].str.len().value_counts())
```

```python
mask = (
    cleaned['phone'].notna() &
    (cleaned['phone'] != '') &
    ~cleaned['phone'].str.startswith('+')
)

cleaned.loc[mask, 'phone'] = '+' + cleaned.loc[mask, 'phone']
```

```python
print(cleaned['phone'].str.len().value_counts(dropna=False))
```

```python
cleaned['Respons. user'].unique()
```

```python
cleaned['Respons. user'].astype(str).str.strip()
```

```python
cleaned['course_name'] = (
    cleaned['Lead title']
    .combine_first(cleaned['Курс'])
    .combine_first(cleaned['Название пакета'])
    .fillna('not provided')
    .astype(str)
)

cleaned = cleaned.drop(columns  = [
    'Lead title',
    'Курс',
    'Название пакета'
    ]
)
cleaned['course_name'].head(20)
```

```python
cleaned['Lead stage'].unique()
```

```python
cleaned['Lead stage'] = cleaned['Lead stage'].str.replace(r'\s*\(.*?\)', '', regex=True)
```

```python
cleaned['Lead stage'].unique()
```

```python
cleaned['Оплата'].unique()
```

```python
cleaned['Contact respons. user'].unique()
```

```python
cleaned['gclid'].unique()
```

```python
cleaned['COOKIES'].unique()
```

```python
cleaned['Contact`s company'].unique()
```

```python
cleaned['FORMID'].unique()
```

```python
cleaned['Modified by'].unique()
```

```python
cleaned = cleaned.drop(columns =['gclid','COOKIES','Contact`s company'])
```

```python
cleaned['manager'] = cleaned[
    ['Respons. user', 'Contact respons. user','Modified by']
].bfill(axis=1).iloc[:, 0]

cleaned = cleaned.drop(columns=[
    'Respons. user',
    'Contact respons. user',
    'Modified by'
])
```

```python
cleaned['manager'].value_counts()
```

```python
cleaned['manager'].isna().sum()
```

```python
cleaned['utm_referrer'] = (
    cleaned['utm_referrer']
    .combine_first(cleaned['REFERER'])
    .combine_first(cleaned['referrer'])
)

cleaned = cleaned.drop(columns= [
    'REFERER',
    'referrer'
])
```
